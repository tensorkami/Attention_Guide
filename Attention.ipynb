# V

```python
v1 = x1.mm(Wv) # shape (1, Attention_D)
v2 = x2.mm(Wv) # shape (1, Attention_D)
v3 = x3.mm(Wv) # shape (1, Attention_D)
v4 = x4.mm(Wv) # shape (1, Attention_D)
v5 = x5.mm(Wv) # shape (1, Attention_D)
v6 = x6.mm(Wv) # shape (1, Attention_D)
```

## Score

```python
K = torch.cat([k1,k2,k3,k4,k5,k6])  # shape (set_num, Attention_D)
```

```python
score1 = q1.mm(K.T) # shape (1, set_num)
score2 = q2.mm(K.T) # shape (1, set_num)
score3 = q3.mm(K.T) # shape (1, set_num)
score4 = q4.mm(K.T) # shape (1, set_num)
score5 = q5.mm(K.T) # shape (1, set_num)
score6 = q6.mm(K.T) # shape (1, set_num)
```

```python
score1 = torch.softmax(score1,1) # shape (1, set_num)
score2 = torch.softmax(score2,1) # shape (1, set_num)
score3 = torch.softmax(score3,1) # shape (1, set_num)
score4 = torch.softmax(score4,1) # shape (1, set_num)
score5 = torch.softmax(score5,1) # shape (1, set_num)
score6 = torch.softmax(score6,1) # shape (1, set_num)
```

```python
V = torch.cat([v1,v2,v3,v4,v5,v6])  # shape (set_num, Attention_D)
```

```python
y1 = score1.mm(V)  # shape (1, Attention_D)
y2 = score2.mm(V)  # shape (set_num, Attention_D)
y3 = score3.mm(V)  # shape (set_num, Attention_D)
y4 = score4.mm(V)  # shape (set_num, Attention_D)
y5 = score5.mm(V)  # shape (set_num, Attention_D)
y6 = score6.mm(V)  # shape (set_num, Attention_D)
```
